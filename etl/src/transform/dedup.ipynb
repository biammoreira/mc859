```python
import pandas as pd
```

```python
genre_df = pd.read_csv('genres.csv')
```

```python
dedup_genre_df = genre_df.drop_duplicates(keep='first')
```

```python
dedup_genre_df.to_csv('dedup_genres.csv', index=False)
```

```python
actor_df = pd.read_csv('actors.csv')
```

```python
dedup_actor_df = actor_df.drop_duplicates(keep='first')
```

```python
dedup_actor_df.to_csv('dedup_actors.csv', index=False)
```

```python
movie_actor_df = pd.read_csv('movie_actor.csv')
```

```python
movie_actor_dedup_df = movie_actor_df.drop_duplicates(keep='first')
```

```python
movie_actor_dedup_df.to_csv('dedup_movie_actor.csv', index=False)
```
